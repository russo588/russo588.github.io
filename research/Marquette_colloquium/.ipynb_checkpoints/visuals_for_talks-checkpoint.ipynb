```python
import numpy as np
import math
from matplotlib import cm
from numpy.linalg import norm as norm
import matplotlib.pyplot as plt
import scipy as scipy
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import griddata
from scipy.integrate import odeint
from itertools import product 
from numpy import linalg as LA
from joblib import Parallel, delayed
```

```python
mu = .5
def ker(x,y):
    return math.exp(-1/mu * norm(x-y)**2)
def occker(x, traj, time):
    arg = np.array([ker(x, traj[i]) for i in range(len(time))])
    return scipy.integrate.simpson(arg,time)
```

```python
def plot_occ(traj, time ,x_range, y_range, res = 0.1, angle1=0, angle2=0):
    my_dpi = 200
    
    plt.figure()
  
    ax = plt.axes(projection='3d')
    ax.set_facecolor('#343333')


    # Make data.
    [x_min, x_max] = x_range
    [y_min, y_max] = y_range
    X = np.arange(x_min, x_max, res)
    Y = np.arange(y_min, y_max, res)
    X, Y = np.meshgrid(X, Y)
    Z = np.zeros(X.shape)
    w = X.shape[0]
    l = X.shape[1]
    cartprod = product(range(w), range(l))
    Z = np.array(Parallel(n_jobs=4)(delayed(occker)([X[i,j], Y[i,j]], traj, time) for (i,j) in cartprod))
    Z = Z.reshape(X.shape)    
    
    
    ax.axis('off')
    ax.view_init(angle1, angle2)
    plt.plot(traj[:,0],traj[:,1], color = 'white')
    surf = ax.plot_surface(X, Y, Z, cmap = cm.cool, antialiased=True, alpha = 0.74)
    
    plt.savefig('occkernel_spiral.png', dpi = my_dpi, bbox_inches='tight', pad_inches=0)
        
```

```python
import glob
from PIL import Image
  

def make_gif():
    howmany = glob.glob("*.png")
    image_list = ['frame'+str(i)+'.png' for i in range(len(howmany))]
    frames = [Image.open(image) for image in image_list]
    frame_one = frames[0]
    frame_one.save("occkernel.gif", format="GIF", append_images=frames,
               save_all=True, duration=100, loop=0)



```

```python
from PIL import Image
import PIL.ImageOps  

def invert():
    howmany = glob.glob("*.png")
    for i in range(len(howmany)):
        file = 'frame'+str(i)+'.png'
        image = Image.open(file)
        if image.mode == 'RGBA':
            r,g,b,a = image.split()
            rgb_image = Image.merge('RGB', (r,g,b))

            inverted_image = PIL.ImageOps.invert(rgb_image)

            r2,g2,b2 = inverted_image.split()

            final_transparent_image = Image.merge('RGBA', (r2,g2,b2,a))

            final_transparent_image.save(file)
```

```python
def plot_gif(traj, time ,x_range, y_range, res = 0.1):
    my_dpi = 200
    
    plt.figure()
  
    ax = plt.axes(projection='3d')
    ax.set_facecolor('#343333')


    # Make data.
    [x_min, x_max] = x_range
    [y_min, y_max] = y_range
    X = np.arange(x_min, x_max, res)
    Y = np.arange(y_min, y_max, res)
    X, Y = np.meshgrid(X, Y)
    Z = np.zeros(X.shape)
    w = X.shape[0]
    l = X.shape[1]
    cartprod = product(range(w), range(l))
    Z = np.array(Parallel(n_jobs=4)(delayed(occker)([X[i,j], Y[i,j]], traj, time) for (i,j) in cartprod))
    Z = Z.reshape(X.shape)    
    
    for i in range(len(traj)):
        ax.axis('off')
        plt.plot(traj[:,0],traj[:,1], color = 'white')
        plt.plot([1.9, 1.9], [-2, 4], ':',color = 'white')
        ax.scatter(traj[i,0],traj[i,1],0, color = 'white', s=50)
        surf = ax.plot_surface(X, Y, Z, cmap = cm.cool, antialiased=True, alpha = 0.74)
        index = str(i)
        plt.savefig('frame'+index+'.png', dpi = my_dpi, bbox_inches='tight', pad_inches=0)
        plt.cla()
    #invert()
```
